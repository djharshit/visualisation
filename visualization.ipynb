```python
# Importing the modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load the dataset
df = sns.load_dataset(name='iris')
```

```python
# Line Chart
plt.plot(df.petal_width)
plt.show()
```

```python
# Bar Chart
b = df.species.value_counts()
plt.bar(b.index, b.values, color=['red', 'green', 'blue'])
plt.show()
```

```python
# Histogram
plt.hist(df.petal_width)
plt.show()
```
